> 1. Unlabeled Columns, we have to label it.
> 2. Unnecessary Columns. i.e, name columns. 
> 3. Empty rows.
> 4. leading/trailing whitespaces.
> 5. Replce the missing values.

**Removing the 0th row. Becuase, it don't have any use.**

```python
df = df[df.index != 0]

df.head(216)
```

**Removing Name and Roll No columns becuase, it is not related as such to marks.**

```python
df = df.drop(['Name', 'Roll No'], axis=1)
df
```

**Function to check if a value is numeric**

```python
def is_numeric(x):
    return isinstance(x, (int, float))
```

**Counting and removing empty values.**

```python
# Count non-numeric and empty values before removal

non_numeric_count = sum(~df[col].apply(is_numeric).all() for col in df.columns)

empty_string_count = sum(df[col].apply(lambda x: isinstance(x, str) and x.strip() == '').sum() for col in df.columns)

print('non_numeric_count before removal:', non_numeric_count)

print('empty_string_count before removal:', empty_string_count)
```

```python
# Remove non-numeric and empty values

for col in df.columns:
    df[col] = pd.to_numeric(df[col], errors='coerce')


df = df.dropna(subset=df.columns)
```

```python
# Count non-numeric and empty values after removal

non_numeric_count_after = sum(~df[col].apply(is_numeric).all() for col in df.columns)

empty_string_count_after = sum(df[col].apply(lambda x: isinstance(x, str) and x.strip() == '').sum() for col in df.columns)

print('non_numeric_count after removal:', non_numeric_count_after)

print('empty_string_count after removal:', empty_string_count_after)
```

```python
df
```

**Checking and removing leading/trailing whitespaces in the DataFrame**

```python
# Check for leading/trailing whitespaces in the DataFrame

whitespace_issues = {}

for column in df.columns:
    
    has_whitespace = df[column].apply(lambda x: isinstance(x, str) and (x != x.strip())).any()
    
    whitespace_issues[column] = has_whitespace

# Print summary of columns with whitespace issues

for column, has_whitespace in whitespace_issues.items():
    if has_whitespace:
        print(f"Columns with leading/trailing whitespace:- {column}")

# Ensure there are no whitespaces

if not any(whitespace_issues.values()):
    print("No leading or trailing whitespaces found in the dataset.")
```

```python
# Replace empty strings with '0', convert to float, and fill missing values with 0

df = df.replace('', '0').astype(float).fillna(value=0)
```

# Define Target Variable (y) and Feature Variables (X)

```python
y = df['Mid_Term']

X = df[['Quize_01', 'Quize_02', 'Quize_Avg', 'Ass_01', 'Ass_02', 'Ass_Avg']]
```

```python
X.shape, y.shape
```

```python
X.value_counts().sum(), y.value_counts().sum()
```

# Train Test Split

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size = 0.7, 
                                                    random_state = 2529)
```

# Model Evaluation

```python
from sklearn.tree import DecisionTreeRegressor

model = DecisionTreeRegressor()
```

```python
model = model.fit(X_train, y_train)
```

# Prediction

```python
y_pred = model.predict(X_test)
```

# Accuracy

```python
from sklearn.metrics import mean_absolute_error, mean_absolute_percentage_error, r2_score
```

```python
print(mean_absolute_error(y_pred, y_test))
print(mean_absolute_percentage_error(y_pred, y_test))
print(r2_score(y_pred, y_test))

mean_absolute_percentage_error = mean_absolute_percentage_error(y_pred, y_test)
print(f"\nAccuracy: {100-mean_absolute_percentage_error}%")
```
